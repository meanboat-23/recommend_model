```python
osaka = '/content/osaka'
```

```python
all_att_reviews = glob(join(osaka,'*'))
all_att_reviews[:5]
```

```python
all_df = pd.DataFrame()
for path in all_att_reviews:
  df = pd.read_csv(path)
  all_df = pd.concat([all_df,df])
all_df = all_df.reset_index()
```

```python
all_df = all_df.drop(['index'],axis=1)
```

```python
df = pd.read_csv('/content/total_Osaka.csv')
```

```python
df['keyword']
```

```python
na_list = df[df['keyword'].isna() == True].index
```

```python
df_key = []
for i in range(len(df)):
  if i not in na_list:
    df_key.append(eval(df['keyword'][i]))
  else:
    df_key.append('Nan')
```
